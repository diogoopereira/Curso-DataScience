```python
#Recriamos o grafo 4
grafo4 = Graph(directed = False)
grafo4.add_vertices(5)
grafo4.add_edges([(0,1),(1,2),(2,3),(3,4),(4,0),(0,2),(2,1)])
grafo4.add_vertex(5)
grafo4.add_vertex(6)
grafo4.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
grafo4.vs['name'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
```

```python
#Impressao da matriz de adjacencia
print(grafo4.get_adjacency())
```

```python
#Linha
grafo4.get_adjacency()[0,]
```

```python
#Linha e coluna
grafo4.get_adjacency()[0,1]
```

```python
#Estrutura de repetição para percorrer cada vertice, visualizando o nome e o rotulo
for v in grafo4.vs:
  print(v)
```

```python
plot(grafo4, bbox = (0,0,300,300))
```

```python
#Criação do grafo com peso entre as relações
grafo5 = Graph(edges = [(0,1), (2,3), (0,2), (0,3)], directed = True)
grafo5.vs['label'] = ['Fernando', 'Pedro', 'Jose', 'Antonio']
grafo5.vs['peso'] = [40,30,30,25]
print(grafo5)
```

```python
#Percorrer os vertices para visualizar os pesos
for v in grafo5.vs:
  print(v)
grafo5.vs[0]
```

```python
#Definição do tipo de amizade e do peso das relações
grafo5.es['TipoAmizade'] = ['Amigo', 'Inimigo', 'Inimigo', 'Amigo']
grafo5.es['weight'] = [1,2,1,3]
print(grafo5)
```

```python
#Percorrer os vertices, tipo de amizade
for e in grafo5.es:
  print(e)
```

```python
#Propriedades e valores de uma posição
grafo5.vs[0]
```

```python
#Tipos de amizade
grafo5.es['TipoAmizade']
```
