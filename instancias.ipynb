```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn import datasets
from scipy import stats
```

```python
iristemp = pd.read_csv('iris.csv')
iristemp.head()
```

```python
iris = datasets.load_iris()
stats.describe(iris.data)
```

```python
#classe e dados
#iris.target
iris.data
```

```python
#Criação das previsoes (variavel independente - X) e class (variavel dependente - y)
previsores = iris.data
classe = iris.target
```

```python
#Divisao da base de dados entre treino e teste(70% para treinamento e 30% para teste)
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores,
                                                                  classe,
                                                                  test_size = 0.3,
                                                                  random_state = 0)
len(X_treinamento)
```

```python
#Criação do modelo, treinamento,
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_treinamento, y_treinamento)
```

```python
#obtençãop das previsoes
previsoes = knn.predict(X_teste)
previsoes
```

```python
#obtenção da matriz de confusão
confusao = confusion_matrix(y_teste, previsoes)
confusao
```

```python
#taxas de acerto e erro
taxa_acerto = accuracy_score(y_teste, previsoes)
taxa_erro = 1 - taxa_acerto
taxa_acerto
```
