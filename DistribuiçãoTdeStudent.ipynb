```python
from scipy.stats import t
```

```python
#Qual a probabilidade de selecionar um cientista de dados e o salario ser menos que R$80 por hora
t.cdf(1.5, 8)
```

```python
#Qual a probabilidade de salario ser maior que 80?
t.sf(1.5, 8)
```

```python
#Somatorio da execução dos dois codigos acima (lado esquerdo + lado direito da distribuição)
t.cdf(1.5, 8) + t.sf(1.5, 8)
```
