```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
#Criação da matriz com os dados e execução do teste
novela = np.array([[19, 6], [42, 32]])
novela
```

```python
#Segundo valor é pvalue
#Valor de p é maior que 0,05 não já evidencias de diferença significativa (hipótese nula): não há diferença significativa
chi2_contingency(novela)
```

```python
novela2 = np.array([[22, 3], [43, 32]])
novela2
```

```python
#Agora valor de p menor que 0,05, podemos rejeitar a hipótese alternativa: há diferença significativa
chi2_contingency(novela2)
```
